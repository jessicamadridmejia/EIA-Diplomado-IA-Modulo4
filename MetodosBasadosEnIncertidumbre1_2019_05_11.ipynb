```
# numpy: libreria para calculos matriciales
# np: alias de numpy
import numpy as np

# matplotib.pyplot: permite crear gráficas
# plt: alias de pyplot
import matplotlib.pyplot as plt #Libreía para graficar




# CONJUNTO BORROSO DISCRETO

# np.arange(): retorna un vector con una secuentcia de números
# Parámetros:
# 1: [0,#)
# 2: [#1,#2)
# 2: [#2,#3) #3 pasa

# Creo el universo del discurso
X = np.arange(1,11) 
print(X)

# np.array(): me permite crear un arreglo

# Guardo los valores de mebresía del conjunto
uA = np.array([0,0.1,0.4,0.8,1,1,0.8,0.4,0.1,0])
print(uA)

# Muestro el tamaño de X y uA
print("El tamaño de X es: "+str(X.size)) #size() tamaño del vector
print("El tamaño de uA es: "+str(uA.size))

# plt.plot(X,Y): graficar X vs Y

# Graficamos el conjunto borroso A
plt.plot(X,uA)
#plt.plot(X,uA,'r')
#plt.plot(X,uA,'r--')
#plt.plot(X,uA,'ro')



plt.figure() # para que las graficas queden separadas




# CONJUNTO BORROSO CONTINUO

# np.linspace(): 
# Parámetros: 
# 1: inicio
# 2: fin (incluye)
# 3: cuantos datos quiero

# Creo el universo del discurso
X = np.linspace(0,10,200)
# Guardo los valores de mebresía del conjunto
uA = np.exp(-(X-5)**2/6) # campana centrada en 5


# Muestro el tamaño de X y uA
print("El tamaño de X es: "+str(X.size)) #size() tamaño del vector
print("El tamaño de uA es: "+str(uA.size))

# Graficamos el conjunto borroso A
plt.plot(X,uA)




# OPERACIONES ENTRE CONJUNTOS BORROSOS

# Creo el universo del discurso
X = np.linspace(0,10,200)

A = np.exp(-(X-4)**2/6) # campana centrada en 4
B = np.exp(-(X-6)**2/6) # campana centrada en 6

print("A: ",A)

# Grafico los conjuntos A y B
plt.figure() # nueva figura para otra grafica
plt.plot(X,A,X,B)


# np.minimum(A,B): retorna un arreglo de la misma dimensión de A y B con los valores minimos

# Grafico la intersección A y B
plt.figure() # nueva figura para otra grafica
plt.plot(X,np.minimum(A,B),'r')

# Grafico la unión A y B
plt.figure() # nueva figura para otra grafica
plt.plot(X,np.maximum(A,B),'r')

# Grafico el complemento
plt.figure() # nueva figura para otra grafica
plt.plot(X,A,X,1-A)

```

```
# Creo el universo del discurso
X = np.linspace(30,250,200)

#Inicializo A y B
A = np.zeros(200)
B = np.zeros(200)

# Genero los valores de membresía de A y B
for i in range(200): #el ciclo debe tener la misma cantidad que el array X
  # Para A
  if (X[i]>75 and X[i]<=200):
    A[i] = (X[i]-75) / 125 # X-75/200-75
  elif (X[i]>200 and X[i]<=250):
    A[i] = (250-X[i]) / 50 # 250-X/250-200
  else:
    A[i]=0
    
  # Para B
  if (X[i]>30 and X[i]<=110):
    B[i] = (X[i]-30) / 80 # X-75/200-75
  elif (X[i]>110 and X[i]<=200):
    B[i] = (200-X[i]) / 90 # 250-X/250-200  
  else:
    B[i]=0

#Grafico los conjuntos    
plt.plot(X,A,X,B)

#Calculo la intercepción
C = np.minimum(A,B)
plt.figure()
plt.plot(X,C) #grafico de la intercepcion
print(np.max(C)) # punto maximo de la intercepción (posibilidad)
print(X[np.argmax(C)]) # punto maximo de la intercepción (universo del discurso)
```
