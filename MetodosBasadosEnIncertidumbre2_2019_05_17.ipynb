```
#Conjunto borroso velocidad alta
X = [0,40,80,100,120]
Y = [1,2,3,4]

#Conjunto borroso accidente grave
A = [0,0.1,0.5,0.8,1]
B = [0.7,0.8,0.9,1]

plt.figure()
plt.plot(X,A)
plt.figure()
plt.plot(Y,B)

# Relación
R = np.zeros((5,4))
print(R)

for f in range(5):
  for c in range(4):
    R[f][c] = min(A[f],B[c])
    
print(R)

plt.figure()
plt.contourf(R) #mapa de calor visto desde arriba
```

```
#conjuto borroso para oferta de libros alta
X = np.linspace(0,500,501)
A = fuzz.trimf(X,[300,500,500])

plt.figure()
plt.plot(X,A)

#conjuto borroso para venta de libros moderada
Y = np.linspace(0,100,101)
B = fuzz.trimf(Y,[30,50,70])

plt.figure()
plt.plot(Y,B)


#Producto cartesiano entre A y B

# Relación
R = np.zeros((501,101))

for f in range(501):
  for c in range(101):
    R[f][c] = min(A[f],B[c])
    




# Librería para gráficas 3D
from mpl_toolkits.mplot3d import Axes3D
#Librería de colores
from matplotlib import cm

#Capturo una gráfica 2D
fig = plt.figure()
# Le agrego un eje más (Eje Z)
ax = Axes3D(fig)
#Creo una malla de valores para la gráfica
x,y = np.meshgrid(X,Y)
#Grafico de la superficie
ax.plot_surface(x,y,R.transpose(),cmap=cm.viridis) # hace la traspuesta de R porque tiene las dimensiones al contrario de x y y

plt.figure()
plt.contourf(R) #mapa de calor visto desde arriba

print(x.shape)
print(y.shape)
print(R.shape)
```

```
# Conjunto borroso cantidad nubes alta
X = np.linspace(0,8,101) # universo
A = fuzz.sigmf(X,6.5,4) # 6.5 centro, 4 pendiente

plt.plot(X,A)

# Conjunto borroso lluvia
Y = np.linspace(0,1,101) # de 0 a 10 con 100 puntos
B = fuzz.gaussmf(Y,0.5,0.2)

plt.figure()
plt.plot(Y,B)

#Producto cartesiano entre A y B

R = np.zeros((101,101))

for f in range(101):
  for c in range(101):
    R[f][c] = min(A[f],B[c])


#Capturo una gráfica 2D
fig = plt.figure()
# Le agrego un eje más (Eje Z)
ax = Axes3D(fig)
#Creo una malla de valores para la gráfica
x,y = np.meshgrid(X,Y)
#Grafico de la superficie
ax.plot_surface(x,y,R.transpose(),cmap=cm.viridis) # hace la traspuesta de R porque tiene las dimensiones al contrario de x y y

plt.figure()
plt.contourf(x,y,R) #mapa de calor visto desde arriba
    
# Calculo la composición
AP = fuzz.sigmf(X,40,-4)
X = np.linspace(0,100,101) # universo
C = fuzz.maxmin_composition(AP.transpose(),R)[0]

plt.figure()
plt.plot(Y,C)
```

```
# Conjunto borroso cantidad nubes alta
X = np.linspace(0,8,201) # universo
A = fuzz.sigmf(X,6.5,4) # 6.5 centro, 4 pendiente

plt.figure()
plt.plot(X,A)

# Conjunto borroso probabilidad de lluvia moderada
Y = np.linspace(0,1,101) # de 0 a 10 con 100 puntos
B = fuzz.gaussmf(Y,0.5,0.2)

plt.figure()
plt.plot(Y,B)

# Conjunto borroso tráfico lento
Z = np.linspace(0,100,501) # de 0 a 10 con 100 puntos
D = fuzz.gaussmf(Z,40,-4)

plt.figure()
plt.plot(Z,D)

#Calculo la relación entre A y B
Rab = fuzz.relation_min(A,B)
                       
                       
#Calculo la relación entre B y D
Rbd = fuzz.relation_min(B,D)

#-----------------------        ERROR            
                        
#Capturo una gráfica 2D
fig = plt.figure()
# Le agrego un eje más (Eje Z)
ax = Axes3D(fig)
#Creo una malla de valores para la gráfica
x,y = np.meshgrid(X,Y)
#Grafico de la superficie
ax.plot_surface(x,y,Rab.transpose(),cmap=cm.viridis) # hace la traspuesta de R porque tiene las dimensiones al contrario de x y y

plt.figure()
plt.contourf(x,y,Rab) #mapa de calor visto desde arriba
    
#-----------------------        
                        
#Capturo una gráfica 2D
fig = plt.figure()
# Le agrego un eje más (Eje Z)
ax = Axes3D(fig)
#Creo una malla de valores para la gráfica
y,z = np.meshgrid(Y,Z)
#Grafico de la superficie
ax.plot_surface(y,z,Rbd.transpose(),cmap=cm.viridis) # hace la traspuesta de R porque tiene las dimensiones al contrario de x y y

plt.figure()
plt.contourf(y,z,Rbd) #mapa de calor visto desde arriba
 
  
  
# Composición entre Rab yRbd
print(Rab.shape)
print(Rbd.shape)

Comp = fuzz.maxmin_composition(Rab,Rbd)

print(Comp.shape)

#Capturo una gráfica 2D
fig = plt.figure()
# Le agrego un eje más (Eje Z)
ax = Axes3D(fig)
#Creo una malla de valores para la gráfica
x,y = np.meshgrid(X,Z)
#Grafico de la superficie
ax.plot_surface(x,y,Comp.transpose(),cmap=cm.viridis) # hace la traspuesta de R porque tiene las dimensiones al contrario de x y y

plt.figure()
plt.contourf(x,y,Comp) #mapa de calor visto desde arriba
 
```
